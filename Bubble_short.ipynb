```python
from itertools import *
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def bubble_short(A, n):
    comparaciones = 0
    intercambios = 0
    aux = 0
    resultados = []
    for k in range(n-1):
        for i in range(n-k-1):
            if A[i] > A[i+1]:
                aux = 1
                comparaciones += 1
                temp = A[i]
                A[i] = A[i+1]
                A[i+1] = temp
                intercambios += 1
            else:
                comparaciones += 1

    resultados.append(comparaciones+intercambios)
    resultados.append(comparaciones)
    resultados.append(intercambios)
    return resultados
```

```python
n = 10
A = list(i for i in range(n))
acum_ins = 0
acum_c = 0
acum_int = 0
long = 0
instrucciones = []
comparaciones = []
intercambios = []
for a in permutations(A):
    acum = bubble_short(list(a), n)
    acum_ins += acum[0]
    instrucciones.append(acum[0])
    acum_c += acum[1]
    comparaciones.append(acum[1])
    acum_int += acum[2]
    intercambios.append(acum[2])
    long += 1
```

```python
print('ARREGLO DE TAMAÑO {}'.format(n))
print('La cantidad de instrucciones promedio es: {}'.format(float(acum_ins / long)))
print('La cantidad de comparaciones promedio es: {}'.format(float(acum_c / long)))
print('La cantidad de intercambios promedio es: {}'.format(float(acum_int / long)))
```

```python
bins = list(sorted(set(instrucciones)))
bins.append(bins[len(bins)-1]+1)

plt.hist(instrucciones, facecolor='blue',bins = bins,alpha=0.5, edgecolor = 'black',  linewidth=1)
plt.xlabel('Número de instrucciones')
plt.ylabel('Frecuencia')
plt.title('Instrucciones')
plt.show()
```

```python
bins = list(sorted(set(comparaciones)))
bins.append(bins[len(bins)-1]+1)
plt.hist(comparaciones, facecolor='green',  bins = bins,alpha=0.5, edgecolor = 'black',  linewidth=1)
plt.xlabel('Número de comparaciones')
plt.ylabel('Frecuencia')
plt.title('Comparaciones')
plt.show()
```

```python
bins = list(sorted(set(intercambios)))
bins.append(bins[len(bins)-1]+1)
plt.hist(intercambios, facecolor='red',  bins = bins,alpha=0.5, edgecolor = 'black',  linewidth=1)
plt.xlabel('Número de intercambios')
plt.ylabel('Frecuencia')
plt.title('Intercambios')
plt.show()
```
